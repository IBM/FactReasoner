### The Context Retriever
In this example, we show how to use the `ContextRetriever` component to retrieve relevant contexts to a given atom.

### Import libraries

```python
import os
from pathlib import Path
import sys
import logging

parent_dir = str(Path(os.getcwd()).resolve().parent)
sys.path.insert(0, parent_dir)

logger = logging.getLogger(__name__)
logging.getLogger("LiteLLM").disabled = True

from src.fact_reasoner.core.retriever import ContextRetriever
```

### Retrieve contexts for a given atom

```python
# Set the query, cache dir and query builder
query = "Lanny Flaherty has appeared in Law & Order."
cache_dir = None # "my_database.db"
query_builder = None # QueryBuilder(model_id="llama-3.3-70b-instruct", backend="rits")

# Create the context retriever
retriever = ContextRetriever(
    top_k=5,
    service_type="google",
    cache_dir=cache_dir,
    query_builder=query_builder
)

contexts = retriever.query(text=query)

print(f"Number of contexts: {len(contexts)}")
for context in contexts:
    print(context)
    print("****" * 20)

print("Done.")
```

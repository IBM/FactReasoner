```python
from typing import cast

MODEL = "llama-3.3-70b-instruct"
RELEVANCE_THRESHOLD = 4.0
CONFIDENCE_THRESHOLD = 2.0
VERSION = cast(str, "qa")
```

```python
from src.fact_reasoner.atom_extractor import AtomExtractor
from src.fact_reasoner.batch_atom_reviser import BatchAtomReviser
from src.fact_reasoner.comprehensiveness import (
    ComprehensivenessPipeline,
    EndToEndFactGraphMiner,
    NliFactGraphMiner,
    QagFactGraphMiner,
)
from src.fact_reasoner.coverage_evaluator import CoverageEvaluator
from src.fact_reasoner.nli_extractor import NLIExtractor
from src.fact_reasoner.qag_processor import QagProcessor
from src.fact_reasoner.qag_utils import compare_quantities_with_units_definition
from src.fact_reasoner.query_context_summarizer import QueryContextSummarizer
from src.fact_reasoner.relevance_estimator import RelevanceEstimator

query_context_summarizer = QueryContextSummarizer(model=MODEL)

# Instantiate the components for the desired comprehensiveness version
if VERSION == "nli":
    atom_extractor = AtomExtractor(model=MODEL, prompt_version="v2")
    atom_reviser = BatchAtomReviser(model=MODEL)
    relevance_estimator = RelevanceEstimator(model=MODEL)
    nli_extractor = NLIExtractor(model=MODEL)
    fact_graph_miner = NliFactGraphMiner(
        relevance_threshold=RELEVANCE_THRESHOLD,
        atom_extractor=atom_extractor,
        atom_reviser=atom_reviser,
        relevance_estimator=relevance_estimator,
        nli_extractor=nli_extractor,
        max_atomization_batch_workers=1,
    )
elif VERSION == "qa":
    qag_processor = QagProcessor(
        model=MODEL,
        comparison_tools=[compare_quantities_with_units_definition],
        inject_prompt_template=True,
    )
    fact_graph_miner = QagFactGraphMiner(
        relevance_threshold=RELEVANCE_THRESHOLD,
        confidence_threshold=CONFIDENCE_THRESHOLD,
        qag_processor=qag_processor,
    )
elif "e2e" in VERSION:
    if VERSION == "e2e-base":
        coverage_evaluator = CoverageEvaluator(model=MODEL, version="base")
    else:
        coverage_evaluator = CoverageEvaluator(model=MODEL, version="few-shot")
    fact_graph_miner = EndToEndFactGraphMiner(coverage_evaluator=coverage_evaluator)
else:
    raise ValueError(f"Unexpected version {VERSION}")

# Initialise the comprehensiveness pipeline
comprehensiveness_pipeline = ComprehensivenessPipeline(
    fact_graph_miner=fact_graph_miner,
    query_context_summarizer=query_context_summarizer,
)
```

```python
from src.fact_reasoner.comprehensiveness import ComprehensivenessResult, QagFieldData


def show_result(
    results: list[ComprehensivenessResult],
    answer_fields: list[str] = [
        "answer_balanced",
        "answer_high_risk",
        "answer_low_risk",
    ],
    context_fields: list[str] = [
        "collapse_warning",
        "continued_circulation",
        "icos_amoc",
        "safe_from_collapse",
    ],
):
    sample_zero = results[0]["sample"]
    print(f"——————————[ Q: {sample_zero['question']} ]——————————")
    print("—————[ Contexts ]—————")
    for i, context_field in enumerate(context_fields):
        print(f"{i}. {sample_zero[context_field]}")
    print()

    qa_field_data = cast(dict[str, QagFieldData], results[0]["field_data"])

    print("—————[ Summarized contexts ]—————")
    for fid, field in qa_field_data.items():
        if field["field_type"] == "context":
            print(f"[{fid}]")
            print(field["summary"])
    print()

    print("—————[ Raw context questions ]—————")
    print("[Raw questions]")
    for fid, field in qa_field_data.items():
        if field["field_type"] == "context":
            print(f"[{fid}]")
            for question in field["questions"]:
                print(f"— {question}")
    print()

    for i, (answer_field, result) in enumerate(zip(answer_fields, results)):
        sample = result["sample"]

        print(f"—————[ Answer {i} ] —————")
        print("[ Answer text ]")
        answer = sample[answer_field]
        print(answer)
        print()

        print("[ Answer questions ]")
        answer_field_data = qa_field_data[answer_field]
        for question in answer_field_data["questions"]:
            print(f"— {question}")
        print()

        if "all_questions_dict" in sample:
            print("[ Refined questions ]")
            sorted_questions = sorted(
                sample["all_questions_dict"].items(),
                key=lambda q: q[1].relevance,
                reverse=True,
            )
            for question, qd in sorted_questions:
                print(f"{question} [{qd.relevance:.2f}]")
            print()

        print("[ Answer atoms ]")
        for aid, answer_atom in result["atoms"].items():
            print(f"{aid}: {answer_atom.text}")
        print()

        print("—————[ Relevant context atoms ]—————")
        for cid, context_atom in result["contexts"].items():
            print(f"{cid}: {context_atom.text}")
        print()

        print("[ Comprehensiveness evaluation ]")
        covered_contexts = result["covered_contexts"]
        print(f"Covered contexts: {covered_contexts}")
        print("Covered context explanations:")
        for context in covered_contexts:
            explanation = result["fact_graph"].show_coverage_path(context)
            print(f"- {context}: {explanation}")
        print(f"Uncovered contexts: {result['uncovered_contexts']}")
        print("Context coverage details:")
        for cid, context_atom in result["contexts"].items():
            print(
                f"{cid}: {context_atom.text} {'✅' if cid in covered_contexts else '❌'}"
            )
        comprehensiveness = result["comprehensiveness_score"]
        print(f"Comprehensiveness score: {comprehensiveness:.2f}")
        print(f"Main evaluation result: {result['comprehensiveness_eval_main_score']}")
        print()
```
